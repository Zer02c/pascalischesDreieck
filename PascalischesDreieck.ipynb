# die berechnung der fakulät wird mathematisch dargestellt durch ein ausrufezeichen(!)
## ** für das berechnen der fakultät gilt n!( n ∈ ℕ, n ≠ 0 ) **
###     ** Bsp. für eine fkultät 5! = 2* 3* 4* 5 = 120 **

```python
def f(x: int):
    ret : int = 1
    for i in range(1 ,x+1):
        ret : int = ret*i
    return ret
```

# Formel für den Binominalen Koefizzient

##          ** n Über k **


```python
def nCr(n : int, k : int):
    nf : int = f(n)
    kf : int = f(k)
    return nf/(kf*f(n - k))
```

```python
#example lotto
nCr(49, 6)
```
